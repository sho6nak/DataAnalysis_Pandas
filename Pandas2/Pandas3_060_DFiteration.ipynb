## Do necessary Imports

```python
import pandas as pd
import numpy as np
```

## Create a Data Frame

```python
N=20
df = pd.DataFrame({
   'Date': pd.date_range(start='2016-01-01',periods=N,freq='D'),
   'LineNum': np.linspace(0,stop=N-1,num=N),
   'Fraction': np.random.rand(N),
   'RandFromSet': np.random.choice(['Low','Medium','High'],N).tolist(),
   'RandFromRange': np.random.normal(100, 10, size=(N)).tolist()
   })
df.head(20)
```

## Iterate Column names

```python
for colNms in df:
    print(colNms)
```

## The iteritems() to iterate all values column wise.

```python
for col, value in df.iteritems() :
    print('Column: ', col)
    print(value)
    print('----------------------')
```

## The iterrows() to iterate row wise

```python
for row_index,row in df.iterrows():
    print (row_index)
    print(row)
    print('---------------')
```

## The itertuples() to iterate DF and return tuples

```python
for rows in df.itertuples():
    print(rows)
```
